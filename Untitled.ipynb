```python
import os
import json
import numpy as np
import pandas as pd
from datetime import datetime

def process_file(file_path, folder_id, maze_data):
    try:
        with open(file_path, 'r') as f:
            content = f.read()
    except FileNotFoundError:
        print(f"Error: Could not find the file at {file_path}")
        return None

    try:
        parsed_data = json.loads(content)
    except json.JSONDecodeError:
        print(f"Error: Could not decode the JSON data in {file_path}")
        return None

    errors = [dics['error'] for dics in parsed_data[1] if 'error' in dics]
    errors = [_ if _<180 else 360-_ for _ in errors]
    avg_errors = np.average(errors) if errors else None
    std_errors = np.std(errors) if errors else None
    num_mis = 13-len(errors)
    rotation = [1 if _['correct'] == True else 0 for _ in parsed_data[2] if 'correct' in _]
    if len(rotation) != 20:
        for i in range(20-len(rotation)):
            rotation.append(0)
    avg_rotation = np.average(rotation) if rotation else None

    person_data = [
        folder_id,
        1 if parsed_data[3].get('gender') == 'male' else 0,
        1 if parsed_data[3].get('education') == 'Undergraduate' else 0,
        parsed_data[0].get('score'),
        avg_errors,
        std_errors,
        num_mis,
        avg_rotation
    ]
    person_data.extend([maze_data.get('training'), maze_data.get('test')])
    return person_data
def get_maze_data(folder):
    maze_files = [f for f in os.listdir(folder) if f.startswith('MazeData_')]
    
    data_times = {}
    for file in maze_files:
        date_str = file.split('_')[3:6]
        time_str = file.split('_')[-4:]
        am_pm = time_str[-1].split('.')[0]
        time_str = time_str[:-1] + [am_pm]
        
        date_time_str = '_'.join(date_str + time_str)
        date_time = datetime.strptime(date_time_str, '%m_%d_%Y_%I_%M_%S_%p')
        
        with open(os.path.join(folder, file), 'r') as f:
            for line in reversed(f.readlines()):
                if 'Total time in scene' in line:
                    time_in_scene = float(line.split('was')[-1].strip())
                    data_times[date_time] = time_in_scene
                    break

    sorted_times = sorted(data_times.items())
    maze_data = {}
    if sorted_times:
        maze_data['training'] = sorted_times[0][1]
        if len(sorted_times) > 1:
            maze_data['test'] = sorted_times[1][1]

    return maze_data

def get_folder_id(folder_name):
    return ''.join([ch for ch in folder_name if ch.isdigit()])

# Get the list of all folders in the current working directory
current_directory = os.getcwd()
folders = [os.path.join(current_directory, f) for f in os.listdir(current_directory) if os.path.isdir(os.path.join(current_directory, f))]

# Initialize a list to store all people data
all_people_data = []

# Read the Excel file into a DataFrame
info_df = pd.read_csv('../info.csv')

# Loop through all folders and process 'results.txt' in each folder
for folder in folders:
    folder_name = os.path.basename(folder)
    folder_id = get_folder_id(folder_name)
    file_path = os.path.join(folder, 'results.txt')
    maze_data = get_maze_data(folder)
    people_data = process_file(file_path, folder_id, maze_data)
    if people_data is not None:
        all_people_data.append(people_data)


# Create a DataFrame from all_people_data
columns = ['ID', 'Gender', 'Education', 'Score', 'Average Errors', 'Standard Deviation of Errors', 'Number of missing', 'Average Rotation', 'Training Time', 'Test Time']
df = pd.DataFrame(all_people_data, columns=columns)

# Convert 'ID' column to numeric so that merge can be performed correctly
df['ID'] = pd.to_numeric(df['ID'], errors='coerce')

# Merge the data from the Excel file with the data in df
result_df = pd.merge(df, info_df, on='ID', how='left')

# Save the merged DataFrame to a CSV file

# If you want to print the DataFrame, uncomment the next line
(result_df)
```

```python
import pandas as pd

def create_dataframe_from_file(file_path):
    data = {
        "Entry Number": [],
        "Timestamp": [],
        "Head Position X": [], "Head Position Y": [], "Head Position Z": [],
        "Head Rotation X": [], "Head Rotation Y": [], "Head Rotation Z": [],
        "Hand(L) Position X": [], "Hand(L) Position Y": [], "Hand(L) Position Z": [],
        "Hand(L) Rotation X": [], "Hand(L) Rotation Y": [], "Hand(L) Rotation Z": [],
        "Hand(R) Position X": [], "Hand(R) Position Y": [], "Hand(R) Position Z": [],
        "Hand(R) Rotation X": [], "Hand(R) Rotation Y": [], "Hand(R) Rotation Z": []
    }
    
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith('Entry Number'):
                parts = line.strip().split('|')
                data["Entry Number"].append(int(parts[0].split(':')[1].strip()))
                data["Timestamp"].append(float(parts[1].split(':')[1].strip()))
#                 print(parts,"\n")
                for part in parts[2:]:
                    if ':' in part:
                        key, value = part.split(':')
                        key = key.strip()
                        if "Position" in key or "Rotation" in key:
                            x, y, z = map(float, value.strip(" ()").split(','))
                            data[f"{key} X"].append(x)
                            data[f"{key} Y"].append(y)
                            data[f"{key} Z"].append(z)
                    else:
#                         print(f"Unexpected format: {part}")
                        continue
#     print(data)                
    df = pd.DataFrame(data)

    diff_data = {}

    # Calculating the difference between each two consecutive row values
    for column in df.columns:
        if column not in ["Entry Number", "Timestamp"]:
            diff_data[f"Diff_{column}"] = df[column].diff()
    
    diff_df = pd.DataFrame(diff_data)
    
    return df, diff_df
```

```python
# Step 1: Calculate arctangent of the x rotation of the head over y rotation of the head in main DataFrame (df)
df['Head_Rotation_Arctan_XY'] = np.arctan2(df['Head Rotation X'], df['Head Rotation Y'])

# Step 2: Create a diff_df DataFrame with the differences between consecutive rows
diff_df = df.diff()

# Carry over the timestamp values from the original DataFrame to diff_df (as diff operation will make first row NaN)
diff_df['Timestamp'] = df['Timestamp']

# Step 3: Create a new column representing the 'Timestamp' values rounded to the nearest second
diff_df['Timestamp_Second'] = pd.to_datetime(diff_df['Timestamp'], unit='s').dt.floor('1S')

# Step 4: Group by the rounded 'Timestamp' column and sum the arctangent values within each one-second interval
summed_arctangents_per_second = diff_df.groupby('Timestamp_Second')['Head_Rotation_Arctan_XY'].sum().reset_index()

# Merge the summed values back into the diff_df DataFrame
result_df = pd.merge(diff_df, summed_arctangents_per_second, on='Timestamp_Second', suffixes=('', '_Sum_Per_Second'))
```
